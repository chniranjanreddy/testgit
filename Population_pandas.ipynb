# Reading population data and analysing the statistics

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_excel ("population_total (1).xlsx",0) #read excel
df1.head()
```

```python
df3=df1.T
df3.head()
```

```python
df3.insert(0,column='year',value=df3.index)
```

```python
df3.head()
```

```python
ax = df3.plot(kind='line',x='year',y='China')
df3.plot(ax=ax,kind='line',x='year',y='India')
```

```python
df4=df1
df4.insert(0,column='country',value=df4.index)
```

```python
df4.head()
```

```python
pop_sort_1800 = df4[1800].sort_values(ascending=False).head(10)
pop_sort_2019 = df4[2019].sort_values(ascending=False).head(10)
pop_sort_2050 = df4[2050].sort_values(ascending=False).head(10)
pop_sort_2050
```

```python
pop_sort_1800.plot(kind='bar')
```

```python
pop_sort_2019.plot(kind='bar')
```
